```python
# https://sisma2016.gov.it/wp-content/uploads/2022/08/Rapporto-a-Giugno-2022_Commissario-Sisma-2016_def.pdf

with open('sisma2016_data/tot2022.csv','w',encoding='utf-8') as o:
    with open('sisma2016_data/tot2022.txt',encoding='utf-8') as f:
        for line in f:
            line = line.strip().replace('Ascoli Piceno','Ascoli_Piceno').replace('Pesaro e Urbino','Pesaro_e_Urbino').split()
            line[2 : -2] = [' '.join(line[2 : -2])]
            
            if line[2] == 'vari':
                continue
            line[-2] = line[-2].replace('.','').replace(',','.')
            line[-1] = line[-1].replace('.','').replace(',','.')
            line = ','.join(line)
            o.write(line+'\n')
print('done')
import pandas as pd
df = pd.read_csv('sisma2016_data/tot2022.csv',na_values=['-'])
df      
```

```python
with open('sisma2016_data/priv2022.csv','w',encoding='utf-8') as o:
    with open('sisma2016_data/priv2022.txt',encoding='utf-8') as f:
        for line in f:
            line = line.strip().replace('Ascoli Piceno','Ascoli_Piceno').replace('Pesaro e Urbino','Pesaro_e_Urbino').split()
            line[2 : -4] = [' '.join(line[2 : -4])]
            
            if line[2] == 'vari':
                continue
            
            line[-4] = line[-4].replace('.','').replace(',','.')
            line[-3] = line[-3].replace('.','').replace(',','.')
            line[-2] = line[-2].replace('.','').replace(',','.')
            line[-1] = line[-1].replace('.','').replace(',','.')
            line = ','.join(line)
            o.write(line+'\n')
print('done')
import pandas as pd
df = pd.read_csv('sisma2016_data/priv2022.csv',na_values=['-'])
df      
```
